## Task 2 - KMeans Clustering

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df['Species'].head()
```

## Data Visualisation

### Usually in unsupervised learning labels are not provided. But as they are provided here we can visualise the approximate final result.

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.scatterplot(x = 'SepalWidthCm', y = 'PetalWidthCm', data = df, hue = 'Species', palette = 'plasma')
```

## Training the model

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(df.drop(['Species'],axis=1))
```

```python
km.labels_
```

## Visualising and comparing expected and actual result

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
sns.scatterplot(ax = ax1,x = 'SepalWidthCm', y = 'PetalWidthCm', data = df, hue = 'Species', palette = 'plasma',legend=False)
ax2.set_title("Original")
sns.scatterplot(ax = ax2, x = 'SepalWidthCm', y = 'PetalWidthCm', data = df, hue = km.labels_, palette = 'plasma',legend=False)
```

### As one can see, the result seems to be almost equal to the expected output!
